```python
from qiskit.quantum_info import Statevector

zero = Statevector([1,0])
print("zero is : ", zero)
```

```python
zero = Statevector.from_int(0, dims=2)
one = Statevector.from_int(1, dims=2)
print("zero is : ", zero)
print("one is : ", one)
```

```python
psi = one.tensor(zero.tensor(zero))
print( "psi is", psi)
```

```python
from numpy import sqrt
ghz = 1/sqrt(2)*(zero^zero^zero) + 1/sqrt(2)*(one^one^one)
print(f"ghz is {ghz}")
```

```python
from qiskit.quantum_info import Pauli
Z0Z1 = Pauli("ZZI")
print("Z0Z1 is ", Z0Z1)
print("And its matrix is : ")
print(Z0Z1.to_matrix())
```

```python
print("The sparse representation of Z0Z1 is")
print(Z0Z1.to_matrix(sparse=True))
```

```python
# Book p.103 uses qiskit.opflow.PauliOp, removed in Qiskit 1.0 — use SparsePauliOp instead
from qiskit.quantum_info import SparsePauliOp

H_cut = SparsePauliOp("ZZI") + SparsePauliOp("ZIZ")
print("H_cut is : ", H_cut)

print("Sparse representation")
# Book p.103 calls .to_spmatrix() — does not exist on SparsePauliOp; use .to_matrix(sparse=True) instead
print(H_cut.to_matrix(sparse=True))
```

```python
# Version mismatch: book uses qiskit.opflow (I, Z as operator objects with ^ for tensor product),
# removed in Qiskit 1.0. SparsePauliOp is the replacement; ^ is still supported on it.
from qiskit.quantum_info import SparsePauliOp

I_op = SparsePauliOp("I")
Z_op = SparsePauliOp("Z")

H_cut = (Z_op ^ Z_op ^ I_op) + (Z_op ^ I_op ^ Z_op)
print("H_cut is : ", H_cut)

print("Sparse representation")
print(H_cut.to_matrix(sparse=True))
```

```python
import itertools

expectation_values = {}

for bits in itertools.product([0, 1], repeat=3):
    psi = zero if bits[0] == 0 else one
    for bit in bits[1:] :
        if bit == 0 :
            qubit = zero
        else : 
            qubit = one
        psi = psi^qubit
    expectation_value = psi.expectation_value(H_cut)
    expectation_values[bits] = expectation_value

print("--- Expectation Values ---")
for bits, val in expectation_values.items():
    # Transforme (0, 1, 0) en "010"
    bit_str = "".join(map(str, bits))
    
    # .real élimine le "+0j" encombrant
    print(f"|{bit_str}⟩ : {val.real:6.2f}")
```
