```python
# A partir de Page 135 - 4.3.1 : Constrained quadratic models in Ocean
x0 = dimod.Binary("x0")
x1 = dimod.Binary("x1")
x2 = dimod.Binary("x2")

blp = dimod.ConstrainedQuadraticModel()
blp.set_objective(-5*x0+3*x1-2*x2)
blp.add_constraint(x0 + x2 <= 1, "First Costraint")
blp.add_constraint(3*x0 - x1 + 3*x2 <= 4, "Second Costraint")

print(blp.variables)
print(blp.objective)
print(blp.constraints)
```

```python
sample1 = {"x0":1, "x1":1, "x2":1}
print("The assignement is : ", sample1)
print("Its cost is : ", blp.objective.energy(sample1))
print("Is it feasible: ", blp.check_feasible(sample1))
print("The violations of the constraints are : ")
print(blp.violations(sample1))

```

```python
solver = dimod.ExactCQMSolver()
solution = solver.sample_cqm(blp)
print("The list of assignemnet is : ")
print(solution)
```

```python
feasible_sols = solution.filter(lambda s: s.is_feasible)
feasible_sols.first
```

```python
from dwave.samplers import SimulatedAnnealingSampler
# Page 141 - Constrained Quantum Problems to Unconstrained

y0, y1 = dimod.Binaries(["y0","y1"])
cqm  = dimod.ConstrainedQuadraticModel()
cqm.set_objective(-2*y0-3*y1)
cqm.add_constraint(y0 + 2 * y1 <= 2)

qubo, invert = dimod.cqm_to_bqm(cqm, lagrange_multiplier = 5)
print(qubo)

sampler = SimulatedAnnealingSampler()
result = sampler.sample(qubo, num_reads = 10)

print("The solutions that we obtained are : ")
print(result)
```

```python
samples = []
occurences = []

for s in result.data():
    samples.append(invert(s.sample))
    occurences.append(s.num_occurrences)
    sampleset = dimod.SampleSet.from_samples_cqm(samples, cqm, num_occurences=occurences)

print("The solutions to the original problem are")
print(sampleset)
```

```python
# Je filtre les solution impossible et fusionnent celles qui ne diffèrent que pour la valeur des variables slack
final_sols = sampleset.filter(lambda s: s.is_feasible)
final_sols = final_sols.aggregate()
print(final_sols)
```

```python
#Page 165 - Classical Solver
# SteepestDescentSolver

# 'greedy' is the import name for the dwave-greedy package (pip install dwave-greedy)
import greedy
import dimod

J = {(0,1):1, (1,2):1, (2,3):1, (3,0):1}
h = {}
problem = dimod.BinaryQuadraticModel(h, J, 0.0, dimod.SPIN)

solver = greedy.SteepestDescentSolver()
solution = solver.sample(problem, num_reads = 10)
print(solution.aggregate())
```

```python
#TuboSolver - Local search algo with memory and which tries to avoid local minima
# 'tabu' is the import name for the dwave-tabu package (pip install dwave-tabu)
import tabu

solver = tabu.TabuSampler()
solution = solver.sample(problem, num_reads = 10)
print(solution.aggregate())
```

```python
#  SimulatedAnnealingSampler - Local Search algo - tries to move to lower energy solutions.

#  ! pip install dwave-neal

import neal

solver = neal.SimulatedAnnealingSampler()
solution = solver.sample(problem, num_reads = 10)
print(solution.aggregate())
```
